# $\sigma$ vs. $E$ plot

For the Hgamma couling 


[Link](https://hep-fcc.github.io/FCCeePhysicsPerformance/) to lumi and energies and FCC-ee

[MadGraph config](https://github.com/saracreates/mg_configs/blob/main/madgraph_configs/Hgamma_coupling/all_runs.dat) for this data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from cycler import cycler
import matplotlib

# Set the color cycle
#plt.rc('axes', prop_cycle=cycler('color', ['#73b06f', '#e6701b', '#007090', '#c7463d', '#ffad08', '#6f73b0', '#56b4e9', '#4c6e3d', '#c47ca2', '#f5f575']))
# https://coolors.co/542344-f68e5f-e5e059-007ea7-98e2c6
plt.rc('axes', prop_cycle=cycler('color', ['#73b06f', '#e6701b', '#007090', '#e7298a', '#802392', '#ffad08', '#56b4e9']))

# Configure Matplotlib to use LaTeX fonts
matplotlib.rcParams["text.usetex"] = True
matplotlib.rcParams["font.family"] = "serif"

# Set font sizes
matplotlib.rcParams.update(
    {
        "font.size": 16,  # General font size
    }
)
```

```python
file_path = "/afs/cern.ch/work/s/saaumill/public/mg_configs/madgraph_configs/Hgamma_coupling/xsec-vs-energy.csv"
data = np.loadtxt(file_path, delimiter=",", skiprows=1)  # Skip header
e = data[:, 0]  # First column
xsec = data[:, 1]  # Second column
```

```python
plt.figure(figsize=(6, 4.5))
plt.plot(e, xsec*10**(6), label=r"$e^+e^- \rightarrow H \gamma$", linewidth=2,)
plt.grid()
plt.ylabel("$\sigma$ (ab)")
plt.xlabel("$\sqrt{s}$ (GeV)")

plt.vlines([160, 240, 365], 0, 88, colors='grey', linestyles='dashed', label=r"$\sqrt{s}$  at FCC-ee")
plt.legend()

# Add custom ticks
xticks = [160, 240, 365]
plt.xticks(list(plt.xticks()[0]) + xticks)  # Keep existing ticks and add custom ones

# Change only the color of the specific ticks
for tick in xticks:
    plt.gca().get_xticklines()[list(plt.xticks()[0]).index(tick)].set_color('grey')
    plt.gca().get_xticklabels()[list(plt.xticks()[0]).index(tick)].set_color('grey')

plt.xlim(130, 506)
#plt.tight_layout()

# Adding text outside the plot in the upper right corner
plt.text(plt.xlim()[1], plt.ylim()[1], "MadGraph5_aMC_v3.5.7", 
         verticalalignment='bottom', horizontalalignment='right', fontsize=16)



plt.tight_layout()
plt.savefig("./plots/xsec-vs-energy.pdf")
plt.show()
```
